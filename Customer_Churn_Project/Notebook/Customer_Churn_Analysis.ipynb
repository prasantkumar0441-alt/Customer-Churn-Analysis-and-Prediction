#Task 4 - Churn Prediction Model

```python
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)
```

```python
lr = LogisticRegression(
    max_iter=1000
)

lr.fit(
    X_train,
    y_train
)
```

```python
lr_pred = lr.predict(
    X_test
)
```

```python
print(
    "Accuracy:",
    accuracy_score(
        y_test,
        lr_pred
    )
)

print(
    "Precision:",
    precision_score(
        y_test,
        lr_pred
    )
)

print(
    "Recall:",
    recall_score(
        y_test,
        lr_pred
    )
)

print(
    "F1 Score:",
    f1_score(
        y_test,
        lr_pred
    )
)
```

```python
print(
    classification_report(
        y_test,
        lr_pred
    )
)
```

```python
dt = DecisionTreeClassifier(
    random_state=42
)

dt.fit(
    X_train,
    y_train
)
```

```python
dt_pred = dt.predict(
    X_test
)
```

```python
print(
    "Accuracy:",
    accuracy_score(
        y_test,
        dt_pred
    )
)

print(
    "Precision:",
    precision_score(
        y_test,
        dt_pred
    )
)

print(
    "Recall:",
    recall_score(
        y_test,
        dt_pred
    )
)

print(
    "F1 Score:",
    f1_score(
        y_test,
        dt_pred
    )
)
```

```python
dt_tuned = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)

dt_tuned.fit(
    X_train,
    y_train
)
```

```python
dt_tuned_pred = dt_tuned.predict(
    X_test
)
```

```python
print(
    "Accuracy:",
    accuracy_score(
        y_test,
        dt_tuned_pred
    )
)

print(
    "Precision:",
    precision_score(
        y_test,
        dt_tuned_pred
    )
)

print(
    "Recall:",
    recall_score(
        y_test,
        dt_tuned_pred
    )
)

print(
    "F1 Score:",
    f1_score(
        y_test,
        dt_tuned_pred
    )
)
```

#Task 5 - Model Evaluation & Interpretation

```python
from sklearn.metrics import (
    roc_curve,
    roc_auc_score
)
```

```python
lr_probs = lr.predict_proba(
    X_test
)[:,1]
```

```python
auc = roc_auc_score(
    y_test,
    lr_probs
)

print(
    "AUC Score:",
    auc
)
```

```python
fpr, tpr, thresholds = roc_curve(
    y_test,
    lr_probs
)

plt.figure(figsize=(8,5))

plt.plot(
    fpr,
    tpr,
    label=f"AUC = {auc:.3f}"
)

plt.plot(
    [0,1],
    [0,1],
    linestyle='--'
)

plt.xlabel(
    "False Positive Rate"
)

plt.ylabel(
    "True Positive Rate"
)

plt.title(
    "ROC Curve"
)

plt.legend()

plt.show()
```

# Task 6 - Business Recommendations

Business Recommendations
1. Encourage Customers to Switch to Long-Term Contracts

The analysis revealed that customers with month-to-month contracts exhibited the highest churn rate. The company should encourage these customers to move to one-year or two-year contracts by offering discounts, loyalty rewards, or exclusive benefits. This strategy can improve customer retention and reduce churn.

2. Promote Automatic Payment Methods

Customers using electronic check payments showed a higher tendency to churn. The company should encourage customers to adopt automatic payment methods such as credit cards or bank transfers by providing incentives, cashback offers, or discounts. This can enhance customer commitment and reduce churn.

3. Strengthen Customer Onboarding Programs

New customers were identified as the segment with the highest churn rate. To improve retention, the company should implement effective onboarding programs, welcome offers, regular follow-ups, and proactive customer support during the initial months of service.

4. Retain High-Value Customers

The analysis identified 578 high-value customers who had churned. These customers contribute significantly to company revenue and should receive special attention through personalized offers, loyalty programs, premium support services, and targeted retention campaigns.

5. Improve Fiber Optic Service Experience

Fiber optic internet service emerged as one of the strongest factors associated with customer churn. The company should investigate potential issues related to pricing, service quality, network performance, and customer satisfaction. Regular feedback collection and service improvements can help reduce churn among fiber optic users.

6. Develop Predictive Retention Strategies

The Logistic Regression model achieved an accuracy of 82.11% and an AUC score of 86.23%, demonstrating strong predictive capability. The company can use this model to identify customers who are likely to churn and proactively engage them with retention offers before they leave.

## Conclusion

The project successfully analyzed customer churn patterns and developed predictive models to identify customers at risk of leaving the company.

Key findings revealed that month-to-month customers, electronic check users, and fiber optic internet users exhibited higher churn tendencies.

Among the evaluated models, Logistic Regression achieved the best performance with an accuracy of 82.11% and an AUC score of 86.23%.

Business recommendations were proposed to improve customer retention, reduce churn, and increase long-term revenue.
